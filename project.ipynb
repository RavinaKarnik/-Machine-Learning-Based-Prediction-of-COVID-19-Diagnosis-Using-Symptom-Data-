**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**



```python
df['total_bedrooms'].isna() #check whether any NaN value is present in column total_bedrooms
len(df['total_bedrooms'])
def fun(total_bedrooms):   #create function which checks all above information and conditions
  if total_bedrooms <= 10 :
      return 'small'
  elif total_bedrooms < 1000:
    return 'medium'
  else:
    return 'large'
n=len(df['total_bedrooms'])


df['total_bedroom_size']=df['total_bedrooms'].apply(fun) #this function is used to apply conditions and make new dataset
df




```
