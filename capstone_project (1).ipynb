```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('corona_tested_006.csv')
```

```python
data
```

```python
data.columns
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop(['Test_date','Ind_ID', 'Sex','Known_contact','Age_60_above'],axis=1,inplace=True)
```

```python
for column in data.columns:
  print(f'{column}---{data[column].unique()}')
```

```python
data=data.replace({'TRUE': 1, 'FALSE': 0, 'None': np.NaN, False: 0, True: 1})
```

```python
data
```

```python
for column in data.columns:
  print(f'{column}---{data[column].unique()}')
```

```python
data.dropna()
```

```python
data.isnull().sum()
```

```python
data= data.replace(to_replace='None', value=np.nan).dropna()
```

```python
data.isnull().sum()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(),annot=True,cmap='Purples')
plt.show()
```

```python
X=data.drop('Corona',axis=1)
```

```python
X
```

```python
data['Corona']=data['Corona'].map({'negative':2,'positive':1,'other':0})
```

```python
y=data['Corona']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec_tree = DecisionTreeClassifier()
```

```python
dec_tree.fit(X_train,y_train)
```

```python
y_pred = dec_tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
y_pred=log_reg.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
rfc.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_train, y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
